<h2>Visualization</h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>Creating Pariplot</h2>

```python
sns.pairplot(da)
plt.show()
print("\n The pairplot reveals visible class separations and some clustering, especially between Fare, Age, and Survived.")
```

<h2>Creating Heatmap</h2>

```python
numerical_features = da.select_dtypes(include=['number'])
correlation_matrix = numerical_features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
print("\n Fare shows a moderate positive correlation with Pclass and Survived, while Age and other numerical features have weak correlations.")
```

<h2>Creating Countplot</h2>

```python
sns.countplot(x='Sex', data=da)
plt.title('Distribution of Sex')
plt.show()
print("\n There were more male passengers than female passengers on board.")
```

```python
sns.countplot(x='Pclass', data=da)
plt.title('Distribution of Pclass')
plt.show()
print("\n The majority of passengers were in 3rd class, followed by 1st and then 2nd class.")
```

```python
sns.countplot(x='Embarked', data=da)
plt.title('Distribution of Embarked')
plt.show()
print("\n Most passengers embarked from port 'S', followed by 'C' and then 'Q'.")
```

<h2>Creating Barplot</h2>

```python
sns.barplot(x='Survived', y='Fare', data=da)
plt.title('Average Fare vs Survival')
plt.show()
print("\n Survivors paid a higher average fare compared to non-survivors, suggesting a possible link between fare and survival.")
```

<h2>Creating Histoplot</h2>

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=da, x='Age', bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
print("\n The age distribution is right-skewed, with a higher concentration of passengers in the 20–30 age range.")
```

<h2>Creating Boxplot</h2>

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=da, x='Pclass', y='Age')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

print("\n Passengers in 1st class tend to be older on average compared to those in 2nd and 3rd class.")
```

<h2>Creating Scatterplot</h2>

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=da, x='Age', y='Fare', hue='Survived')
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()


print("\n Observation: 'Fare' and 'Pclass' show a noticeable negative correlation, and 'SibSp' and 'Parch' have a moderate positive correlation, suggesting family size may impact survival.")
```
